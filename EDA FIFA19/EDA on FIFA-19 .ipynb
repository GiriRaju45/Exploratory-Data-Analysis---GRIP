### Analysis on the age of the players

```python
def print_table(header, content, col_widths):
    n_cols = len(header)
    table_length = sum(col_widths) + n_cols +1
    table_format = "|" +" ".join(f"{{:^{w}}}|" for w in col_widths)
    print( "*"*table_length, table_format.format(*header), "*"*table_length, sep = "\n")
    for row in content:
        print(table_format.format(*row))
    print("-"*table_length)
```

```python
age_dict = f_d.Age.value_counts().to_dict()
```

### Players value

```python
# handle all the players
def preprocess_value(x):
    x = str(x).replace('€', '')
    if('M' in str(x)):
        x = str(x).replace('M', '')
        x = float(x) * 1000000
    elif('K' in str(x)):
        x = str(x).replace('K', '')
        x = float(x) * 1000
    return float(x)

f_d['Value'] = f_d['Value'].apply(preprocess_value)
```

```python
#The value has some non numeric mark so I extract rigth value
def value_to_int(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1:]

        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value

f_d['Value_float'] = f_d['Value'].apply(value_to_int)
```

```python
# relAation potential and age with respected value of players
value = f_d.Value_float
sns.relplot(x="Age", y="Potential", hue=value/100000, 
            sizes=(40, 400), alpha=.5,palette = 'deep',
            height=6, data= f_d)
```
